# Entrenamiento
Ahora, en esta sección, entrenaré un modelo de aprendizaje automático para predecir las ventas futuras de un producto. Pero antes de entrenar el modelo, voy a  dividir  los datos en conjuntos de entrenamiento y de prueba:

-------------------------------------------------------------------------------
Training
Now, in this section, I will train a machine learning model to predict future product sales. But before training the model, I will split the data into training and testing sets:

```python
x = np.array(data.drop(["Sales"],axis = 1))

y = np.array(data["Sales"]) #variable objetivo

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```

# Evaluacion del Modelo

El objetivo de evaluar un modelo de regresión lineal es determinar su capacidad para predecir de manera precisa las respuestas de un fenómeno lineal. Se busca medir la calidad de las predicciones, identificar posibles errores y ajustar el modelo para mejorar su capacidad predictiva en datos no vistos. Esto se logra mediante métricas como el error cuadrático medio y la validación cruzada, garantizando que el modelo sea robusto y generalice bien a nuevas observaciones.

--------------------------------------------------------------------------------------


The goal of evaluating a linear regression model is to assess its ability to accurately predict responses in a linear phenomenon. The objective is to measure the quality of predictions, identify potential errors, and fine-tune the model to enhance its predictive capability on unseen data. This is achieved through metrics such as mean squared error and cross-validation, ensuring that the model is robust and generalizes well to new observations.

```python
print(model.score(X_test, y_test))
```

# R^2
Descripcion
* R^2 = 0: El modelo no explica ninguna variabilidad en la variable dependiente.
* R^2 = 1: El modelo explica toda la variabilidad en la variable dependiente.

En este  caso, un R^2 del 90.5% sugiere que el modelo está capturando y explicando una proporción significativa de la variabilidad en tus datos. 

------------------------------------------------------------------------

Description

* R^2 = 0: The model explains no variability in the dependent variable.
* R^2 = 1: The model explains all the variability in the dependent variable.

In this case, an R^2 of 90.5% suggests that the model is capturing and explaining a significant proportion of the variability in your data.

```python
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(mse, rmse)
```

# Error Cuadrático Medio (MSE):

Es la media de los errores al cuadrado entre las predicciones y los valores reales.
Cuanto menor sea el MSE, mejor será el rendimiento del modelo.
MSE penaliza fuertemente los errores grandes.
El MSE es aproximadamente 2.91.

# Raíz del Error Cuadrático Medio (RMSE):

Es la raíz cuadrada del MSE y tiene la misma unidad que la variable de respuesta.
Proporciona una interpretación más intuitiva porque está en la misma escala que la variable de respuesta.
Cuanto menor sea el RMSE, mejor será el rendimiento del modelo.
El RMSE es aproximadamente 1.71.

Un RMSE de 1.71 en particular indica que, en promedio, las predicciones están desviadas por alrededor de 1.71 unidades en la misma escala que la variable de respuesta.

----------------------------------------------------------------------------
* Mean Squared Error (MSE):
It is the average of the squared errors between predictions and actual values. The lower the MSE, the better the model's performance. MSE strongly penalizes large errors. The MSE is approximately 2.91.

* Root Mean Squared Error (RMSE):
It is the square root of MSE and has the same unit as the response variable. It provides a more intuitive interpretation because it is on the same scale as the response variable. The lower the RMSE, the better the model's performance. The RMSE is approximately 1.71.

A specific RMSE of 1.71 indicates that, on average, predictions deviate by around 1.71 units on the same scale as the response variable.

```python
residuals = y_test - y_pred
print(residuals.mean())
```

# Valor Residual Promedio

Un valor negativo indica que, en promedio, el modelo tiende a subestimar las unidades vendidas en comparación con los valores reales. En este caso, el valor de -0.3985 implica que, en promedio, el modelo predice aproximadamente un 0.4% menos unidades de producto vendidas de lo que realmente se observa.

------------------------------------------------------------------------
Average Residual Value
A negative value indicates that, on average, the model tends to underestimate the units sold compared to the actual values. In this case, the value of -0.3985 implies that, on average, the model predicts approximately 0.4% fewer units of the product sold than what is actually observed.

```python

scores = cross_val_score(model, x, y, cv=5)  # 5-fold cross-validation

print ("Result:",scores)
print("Mean: ", np.mean(scores))
print("Std. Dev.: ", np.std(scores))
print("95% CI: ", np.percentile(scores, [2.5,97.5]))
```

La desviación estándar de aproximadamente 0.0314 indica que hay una pequeña variabilidad en los resultados entre los pliegues.

El intervalo de confianza del 95% proporciona una estimación de la variabilidad de los resultados y sugiere que se puede tener confianza en que el rendimiento del modelo está dentro de este rango.

Cross-Validation:
In this case, the model performs quite consistently across different folds, as the scores do not vary significantly, and the confidence interval is narrow.
